```python
qa = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)

context = "Transformers are powerful models for NLP tasks like QA and classification."

result = qa(
    question="What are transformers used for?",
    context=context
)

print(result)
```
