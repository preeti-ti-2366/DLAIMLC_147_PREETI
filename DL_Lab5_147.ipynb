```python
#LSTM Text Classification
import tensorflow as tf
from tensorflow.keras import layers
# Load IMDB dataset

vocab_size = 10000
max_len = 200

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.imdb.load_data(num_words=vocab_size)

# Padding sequences
x_train = tf.keras.preprocessing.sequence.pad_sequences(x_train, maxlen=max_len)
x_test = tf.keras.preprocessing.sequence.pad_sequences(x_test, maxlen=max_len)

model = tf.keras.Sequential([
    layers.Embedding(10000, 128),
    layers.LSTM(64),
    layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=3, batch_size=64)
```

```python
#Seq2Seq Model (Encoder-Decoder)
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense

latent_dim = 256
```

```python
# Dummy data (toy example)
encoder_input_data = np.random.random((1000, 10, 20))
decoder_input_data = np.random.random((1000, 10, 20))
decoder_target_data = np.random.random((1000, 10, 20))
```

```python
# Encoder
encoder_inputs = Input(shape=(None, 20))
encoder = LSTM(latent_dim, return_state=True)
_, state_h, state_c = encoder(encoder_inputs)
encoder_states = [state_h, state_c]
```

```python
# Decoder
decoder_inputs = Input(shape=(None, 20))
decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_inputs, initial_state=encoder_states)
decoder_dense = Dense(20, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)

model = Model([encoder_inputs, decoder_inputs], decoder_outputs)
model.compile(optimizer='adam', loss='categorical_crossentropy')

model.fit([encoder_input_data, decoder_input_data], decoder_target_data,
          batch_size=64, epochs=3)
```
